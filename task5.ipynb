```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10,6)
```

```python
data = pd.read_csv('train.csv')
```

```python
data.set_index("PassengerId")
```

```python
print("First 5 rows:")
data.head()
```

```python
data.info()
```

```python
data .describe()
```

```python
data.isnull().any()
```

```python
for column in data.columns:
    print(f"\nColumn: {column}")
    
    print(data[column].value_counts(dropna=False))
    print("----"*100)
```

```python
# 6. Univariate Analysis
## Target Variable - Survival Count
sns.countplot(x='Survived', data=data)
plt.title('Survival Count')
plt.show()
```

```python
sns.countplot(x='Sex', data=data)
plt.title('Gender Distribution')
plt.show()
```

```python
## Age Distribution
sns.histplot(data['Age'].dropna(), kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
## Passenger Class Distribution
sns.countplot(x='Pclass', data=data)
plt.title('Passenger Class Distribution')
plt.show()
```

```python
# 7. Bivariate Analysis
## Survival by Gender
sns.countplot(x='Survived', hue='Sex', data=data)
plt.title('Survival by Gender')
plt.show()
```

```python
## Survival by Passenger Class
sns.countplot(x='Survived', hue='Pclass', data=data)
plt.title('Survival by Passenger Class')
plt.show()
```

```python
## Age vs Survival
sns.boxplot(x='Survived', y='Age', data=data)
plt.title('Age vs Survival')
plt.show()
```

```python
# 8. Multivariate Analysis
## Correlation Heatmap
numeric_data = data.select_dtypes(include=[np.number])  # Select only numerical columns
corr = numeric_data.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
## Pairplot (Optional: Focus on selected columns)
sns.pairplot(data[['Survived', 'Pclass', 'Age', 'Fare']])
plt.title('Pairplot')
plt.show()
```
